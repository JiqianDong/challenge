```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import geopandas as gpd
```

```python
aq_values_path = '../data/air-2013.csv'

aq_values = pd.read_csv(aq_values_path,names=['SquareId','Time','level','NumberOfReadings'])
```

```python
aq_values['Timestemps'] = pd.to_datetime(aq_values['Time'], format='%Y-%m-%d %H:%M', errors='ignore')
aq_values.drop(columns=['Time'],inplace=True)
```

```python
# Drop the number of readings, sames useless currently

aq_values.drop(columns=['NumberOfReadings'],inplace=True)
```

```python
aq_values['Timestemps_day'] = aq_values['Timestemps'].dt.floor('d')

aq_values.drop(columns=['Timestemps'],inplace=True)
```

```python
aq_values = aq_values.groupby(['Timestemps_day','SquareId']).mean().unstack()
```
