```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import geopandas as gpd
```

```python
p_values_path = '../data/precipitation-trentino.csv'
p_availability_path = '../data/precipitation-trentino-data-availability.csv'

p_values = pd.read_csv(p_values_path,names=['Time','SquareId','level'])
p_availability = pd.read_csv(p_availability_path,names=['Time','Availability'])
```

```python
p_values['Timestemps'] = pd.to_datetime(p_values['Time'], format='%Y%m%d%H%M', errors='ignore')
p_values.drop(columns=['Time'],inplace=True)

p_availability['Timestemps'] = pd.to_datetime(p_availability['Time'], format='%Y%m%d%H%M', errors='ignore')
p_availability.drop(columns=['Time'],inplace=True)
```

```python
p_values.head()
```

```python
# If availability is True, the perciption level at that time is 0

p_values = p_values.groupby(['Timestemps','SquareId']).sum().unstack(level=1).fillna(0)

p_values.shape
```

```python
# Check if consistent

p_availability.loc[p_availability['Availability'] == True]['Timestemps'].count()
```

```python
p_values.columns = p_values.columns.droplevel(0)
```
